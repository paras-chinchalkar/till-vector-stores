```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.docstore.document import Document

# ----- PDF Loader Workflow -----
# Load the PDF file
pdf_loader = PyPDFLoader('dl-curriculum.pdf')
pdf_docs = pdf_loader.load()

# Character splitter configuration
splitter = CharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=0,
    separator=' '  # using space for better chunking
)

# Split PDF documents
pdf_chunks = splitter.split_documents(pdf_docs)

# Print second chunk from PDF
print("PDF Chunk Example:\n", pdf_chunks[1].page_content)
print("-" * 40)


# ----- Raw Text Workflow -----
raw_text = """When shadows dance across the windowpane, the clock ticks louder than it should.
Outside, a storm gathers, waiting to interrupt the rhythm. Inside, silence stretches thin like
overworked canvas. Thoughts loop and unravel, tangled in fluorescent hum. A cat blinks slowly,
then disappears behind the couch. The kettle whistles but no one moves. Pages of half-finished
poems lie scattered like fallen leaves."""

# Wrap raw text in a Document
text_doc = [Document(page_content=raw_text)]

# Split raw text using the same splitter
text_chunks = splitter.split_documents(text_doc)

# Print first chunk from raw text
print("Raw Text Chunk Example:\n", text_chunks[0].page_content)
```
